```python
#visualize the cities w/ the most complaint types
df_service_requests_nyc['City'].value_counts().plot(kind='barh',
                                                              alpha=0.5,
                                                              figsize=(10,10))
plt.title('frequency of complaint types by city')
plt.xlabel('count')
plt.ylabel('cities')
plt.show()
```

```python
#visualize the complaint types
df_service_requests_nyc['Complaint Type'].value_counts().plot(kind='barh',
                                                              alpha=0.5,
                                                              figsize=(10,10))
plt.title('frequency of complaint types')
plt.xlabel('count')
plt.ylabel('complaint types')
plt.show()
```
